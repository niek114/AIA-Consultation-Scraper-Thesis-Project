```python
import pandas as pd

consult = pd.read_csv(
    r"C:\Users\hp EliteBook\Desktop\ai-act-consultation-analysis\data\raw\consultation_2020.csv",
    encoding="cp1252",
    sep=";",
    engine="python",
    dtype=str,
)

letters = pd.read_csv(
    r"C:\Users\hp EliteBook\Desktop\ai-act-consultation-analysis\data\processed\letters_from_pdfs.csv",
    encoding="utf-8-sig",
    sep=";",
    dtype=str,
)
```

```python
consult.columns = [c.replace("\r", " ").replace("\n", " ").strip() for c in consult.columns]
```

```python
letters = letters.drop_duplicates(subset=["reference"], keep="first")

merged = consult.merge(
    letters[["reference", "filename", "text", "quality", "lang_guess"]],
    left_on="Reference",
    right_on="reference",
    how="left",
)

merged = merged.drop(columns=["reference"])
merged = merged.rename(columns={
    "filename": "letter_filename",
    "text": "letter_text",
    "quality": "letter_quality",
    "lang_guess": "letter_lang_guess",
})
```

```python
out_xlsx = r"C:\Users\hp EliteBook\Desktop\ai-act-consultation-analysis\data\processed\consultation_with_letters.xlsx"
merged.to_excel(out_xlsx, index=False)
out_xlsx
```
